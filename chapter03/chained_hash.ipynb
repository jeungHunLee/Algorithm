```python
# 실습 3-5[A]
from __future__ import annotations
import hashlib


class Node:

    def __init__(self, key, value, next):
        self.key = key  # 키
        self.value = value  # 값
        self.next = next  # 뒷쪽 노드 참조


# 실습 3-5 [B]
class ChainedHash:

    def __init__(self, capacity):
        self.capacity = capacity  # 해시 테이블의 크기 지정
        self.table = [None] * self.capacity  # 해시 테이블(리스트) 선언

    # 해시 함수
    def hash_value(self, key):
        if isinstance(key, int):  # key가 int형 객체인지 확인
            return key % self.capacity
        return (int(hashlib.sha256(str(key).encode())).hexdigest(), 16) % self.capacity

    # 실습 3-6 [C]
    # search method
    def search(self, key):
        hash = self.hash_value(key)  # 검색하고자히는 key의 해시값
        p = self.table[hash]  # 노드 지정

        while p is not None:
            if p.key == key:
                return p.value
            p = p.next

        return None

    # add method
    def add(self, key, value):
        hash = self.hash_value(key)  # 추가하고자 하는 key의 해시값
        p = self.table[hash]  # 노드 지정

        while p is not None:
            if p.key == key:
                return False  # 이미 존재하는 key는 추가 실패
            p = p.next

        temp = Node(key, value, self.table[hash])
        self.table[hash] = temp  # 노드 추가
        return True

    # remove method
    def remove(self, key):
        hash = self.hash_value(key)  # 제거하고하 하는 key의 hash값
        p = self.table[hash]  # 노드 설정
        pp = None  # 바로 앞 노드

        while p is not None:
            if p.key == key:
                if pp is None:
                    self.table[hash] = p.next
                else:
                    pp.next = p.next
                return True
            pp = p
            p = p.next
        return False

    # dump method
    def dump(self):
        for i in range(self.capacity):
            p = self.table[i]
            print(i, end="")
            while p is not None:
                print(f' -> {p.key} ({p.value})', end="")
                p = p.next
            print()
```
