### 실습02-08 1000 이하의 소수 나열하기

```python
counter = 0

for n in range(2, 1001):
    check = True
    for i in range(2, n):
        counter += 1
        if n % i == 0:
            check = False
            break
    if check:
        print(n)
    
print(f'나눗셈을 실행한 횟수: {counter}')
```

### 실습02-08 1000 이하의 소수 나열하기 (아리스토텔레스의 체)
위의 코드의 경우 최악의 경우 시간 복잡도 O notation O(N^2)의 시간을 가짐<br>
아리스토텔레스의 체는 최악의 경우에도 O notation O(Nlog(logN))의 시간 복잡도를 보장 하므로 더 효율적임<br>
단순 counter 횟수만 확인하더라도 더 적은 연산으로 소수를 판별할 수 있음을 확인 가능.

```python
counter = 0

check = [True] * 1001

m = int(1000 ** 0.5)    # 31^2 > 1001이므로 30까지 확인해도 층분
for i in range(2, m + 1):
    counter += 1
    if check[i] == True:   # 만약 소수라면
        for j in range(i+i, 1001, i):
            check[j] = False    # 해당하는 배수는 모두 False 처리
            counter += 1
        
for i in range(2, 1001):
    if check[i]:
        print(i)

print(f'나눗셈을 실행한 횟수: {counter}')
```
