### 실습02-10 소수 나열하기 알고리즘 개선

```python
counter = 0    # 연산 횟수
ptr = 0    # 찾은 소수의 개수
prime = [None] * 500    # 소수를 저장할 배열

prime[ptr] = 2    # 2는 소수
ptr += 1

prime[ptr] = 3    # 3도 소수
ptr += 1

for n in range(5, 1001, 2):    # 2를 제외한 짝수는 모두 소수가 아니므로 홀수에 대하여 판별 진행
    i = 1
    while prime[i] * prime[i] <= n:
        counter += 2
        if n % prime[i] == 0:    # 나누어 떨어지면 소수 아님
            break
            
        i += 1
        
    else:    # while문이 break 되지 않고 끝까지 순회 하였을 경우 else문 실행(소수 판정)
        prime[ptr] = n
        ptr += 1
        counter += 1
        
for i in range(ptr):
    print(prime[i])
    
print("연산을 수행한 횟수: " + str(counter))
```
