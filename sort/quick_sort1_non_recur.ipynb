## 비재귀적인 퀵 정렬 구현하기

```python
import import_ipynb
from stack import Stack

def qsort(a, left, right):
    range = Stack(right - left + 1)    # a 배열 크기 많큼 stack 생성
    
    range.push((left, right))
    
    while not range.is_empty():
        pl, pr = left, right = range.pop()
        x = a[(left + right) // 2]    # pivot
        
        while pl <= pr:
            while a[pl] < x: pl += 1
            while a[pr] > x: pr -= 1
                
            if pl <= pr:
                a[pl], a[pr] = a[pr], a[pl]
                pl += 1
                pr -= 1
        if left < pr: range.push((left, pr))    # 왼쪽 그룹 stack에 저장
        if pl < right: range.push((pl, right))    # 오른쪽 그룹 stack에 저장
            
print("퀵 정렬을 수행합니다.")
num = int(input("원소 수를 입력하세요"))
x = [0] * num

for i in range(num):
    x[i] = int(input(f'x[{i}]: '))
                     
qsort(x, 0, len(x) - 1)
                     
print("오름차순으로 정렬했습니다.")
for i in range(num):
    print(f'x[{i}]: {x[i]}')
```
