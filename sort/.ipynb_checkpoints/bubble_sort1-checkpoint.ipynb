## 실습 6-1 버블 정렬 알고리즘 구현하기

```python
def bubble_sort(a):
    n = len(a)
    for i in range(n - 1):
        for j in range(n - 1, i, -1):
            if a[j] < a[j - 1]:
                a[j], a[j - 1] = a[j - 1], a[j]    # 앞에 위치한 요소가 더 클 경우 swap
                
            
print("버블 정렬을 수행합니다.")
num = int(input("원소 수를 입력하세요"))
x = [0] * num

for i in range(num):
    x[i] = int(input(f'x[{i}]: '))
                     
bubble_sort(x)
                     
print("오름차순으로 정렬했습니다.")
for i in range(num):
    print(f'x[{i}]: {x[i]}')
```
