## 실습6-3 버블 정렬 알고리즘 구현하기(알고리즘 개선1)

```python
def bubble_sort(a):
    ccnt = 0    # comparsion count
    scnt = 0    # swap count
    n = len(a)
    for i in range(n - 1):
        exchng = 0    # change count
        print(f'패스 {i + 1}')
        for j in range(n - 1, i, -1):
            for m in range(n - 1):
                print(f'{a[m]:2}' + ('  ' if m != j - 1 else ' +' if a[j - 1] > a[j] else ' -'), end='')
            print(f'{a[n - 1]:2}')
            ccnt += 1    # comparsion count + 1
            
            if a[j - 1] > a[j]:
                a[j - 1], a[j] = a[j], a[j - 1]
                scnt += 1    # swap count +1
                exchng += 1    # change count +1
        
        if exchng == 0:    # 교환이 없으면 이후로는 모두 정렬된 상태이므로 반복문 탈출
            break
    
    for m in range(n - 1):
            print(f'{a[m]:2}', end='  ')
    print(f'{a[n - 1]:2}')
            
    print(f'비교를 {ccnt}번 했습니다.')
    print(f'교환을 {scnt}번 했습니다.')
            
print("버블 정렬을 수행합니다.")
num = int(input("원소 수를 입력하세요"))
x = [0] * num

for i in range(num):
    x[i] = int(input(f'x[{i}]: '))
                     
bubble_sort(x)
                     
print("오름차순으로 정렬했습니다.")
for i in range(num):
    print(f'x[{i}]: {x[i]}')
```
