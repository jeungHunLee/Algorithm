### 실습 5-9 8퀸 알고리즘 구현하기

```python
pos = [0] * 8 
flag_a = [False] * 8    # 행방향 배치 유무
flag_b = [False] * 15    # 대각선 방향(↙↗) 배치 유무
flag_c = [False] * 15    # 대각선 방향(↘↖) 배치 유무

def put():
    for i in range(8):
        print(f'{pos[i]: 2}', end='')
    print()
    
def set(i):
    for j in range(8):
        # 행방향과 대각선 방향(↙↗), 대각선 방향(↘↖)에 모두 퀸이 배치되지 않았을때
        '''
        대각선 방향(↙↗) index: i + j
        대각선 방향(↘↖) index: i - j + 7
        '''
        if(not flag_a[j] and not flag_b[i + j] and not flag_c[i - j + 7]):
            pos[i] = j    # j 행 i 열에 배치
            
            if i == 7:
                put()
                
            else:
                flag_a[j] = flag_b[i + j] = flag_c[i - j + 7] = True    # 배치 처리
                set(i + 1)    # 다음 열에 대하여 배치
                flag_a[j] = flag_b[i + j] = flag_c[i - j + 7] = False    # 다음 경우의 수 확인을 위한 배치 초기화

set(0)
```
