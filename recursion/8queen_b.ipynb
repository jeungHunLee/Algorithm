### 실습 5-8 각 열에 퀸을 1개 배치하는 조합을 재귀적으로 나열하기

```python
# 크기가 8인 리스트로 출력했더니 결과 값이 너무 많아 jupyter notebook이 계속해서 멈춰서 크기가 6인 리스트로 수정했습니다.
pos = [0 for _ in range(6)]

def put():
    for i in range(6):
        print(f'{pos[i]:2}', end='')
    print()
    
def set(i):
    for j in range(6):
        pos[i] = j    # j행 i열
        
        if i == 5:
            put()
        else:
            set(i + 1)
            
set(0)
```
