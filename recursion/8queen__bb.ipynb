### 실습 5-8 행과 열에 퀸을 1개 배치하는 조합을 재귀적으로 나열하기

```python
pos = [0 for _ in range(8)]
flag = [False for _ in range(8)]    # 행에 퀸의 배치 유무

def put():
    for i in range(8):
        print(f'{pos[i]: 2}', end='')
    print()
    
def set(i):
    for j in range(8):
        if not flag[j]:    # j행에 퀸을 배치 하지 않은 경우
            pos[i] = j     # j행 i열에 배치
            
            if i == 7:
                put()
                
            else:
                flag[j] = True     # j행에 배치
                set(i + 1)         # 다음 열에 배치
                flag[j] = False    # set의 재귀 호출이 끝난 경우 False로 초기화(다음 경우의 수를 확인하기 위해)
                
set(0)
```
